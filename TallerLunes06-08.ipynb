```python
import pandas as pd
import sys
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from pandas.tools.plotting import radviz
import ipywidgets as widgets

%matplotlib inline
```

```python
#Unificar los siguientes archivos en un solo dataset: xaa.csv  xab.csv  xac.csv  xad.csv  xae.csv  xaf.csv  xag.csv
data  = pd.read_csv("data/xaa.csv", encoding="latin-1")
data2 = pd.read_csv("data/xab.csv", encoding="latin-1")
data3 = pd.read_csv("data/xac.csv", encoding="latin-1")
data4 = pd.read_csv("data/xad.csv", encoding="latin-1")
data5 = pd.read_csv("data/xae.csv", encoding="latin-1")
data6 = pd.read_csv("data/xaf.csv", encoding="latin-1")
data7 = pd.read_csv("data/xag.csv", encoding="latin-1")
                    
                              
```

```python
data8=pd.concat([data, data2, data3, data4, data5, data6, data7])
```

```python
#Obtener un gráfico que me permita vizualizar los nacimientos en base al atributo tipo_instruccion
plt.figure(figsize=(20,5))
plt.subplot2grid((1,1), (0,0))
data8.niv_inst[data8['anio_insc']=='2017'].value_counts().plot(kind='bar')
plt.title('Nacimientos en base al atributo y tipo de instruccion - Año 2017', fontsize = 20)
plt.show()
```
